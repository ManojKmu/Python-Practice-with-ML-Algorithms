```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plp
```

```python
df=pd.read_csv("J:\data_cleaned.csv")
df.head()
```

```python
df.corr() #statistical Method
```

```python
x=df.drop(['Survived','Fare','Pclass_1','Pclass_3','Sex_female','Sex_male'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.4,random_state=30)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

Predicting the test results for random forest

```python
y_pred=rf.predict(x_test)
y_pred1=dt.predict(x_test)
```

```python
cal=pd.DataFrame(np.c_[y_test,y_pred],columns=["Original","Predicted"])
cal
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score=accuracy_score(y_test,y_pred)
```

```python
accuracy_score # For random forest
```

```python
from sklearn.metrics import accuracy_score   # For Decision Tree
accuracy_score=accuracy_score(y_test,y_pred1)
accuracy_score
```
