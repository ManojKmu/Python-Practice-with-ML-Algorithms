```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("J:\Salary_Data.csv")
df.head()# Defaultly it shows or prints 5 values in the console
```

```python
print(df.shape)# Here shape(it shows rows and columns how many there) can be declared in three ways are 1)df.shape 2)print(df.shape)
```

```python
dt=df["YearsExperience"].corr(df["Salary"])
print(dt)
```

```python
df.isnull().sum()
```

```python
x=df.drop("Salary",axis=True)# If we write like this x=df["YearsExperience"] we get error
y=df["Salary"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=46)# based on test_size it changes the questions(train) and answers(test) values
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
z=regression.predict(x_test)
```

```python
sns.scatterplot(z)
```
