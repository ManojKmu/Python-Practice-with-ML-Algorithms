```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plp
```

```python
df=pd.read_csv("J:\Salary_Data.csv")
df.head()
```

```python
df.corr()
```

```python
x=df.drop(['Salary'],axis=1)
y=df['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3,random_state=30)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegression=LogisticRegression()
LogisticRegression.fit(x_train,y_train)
```

Predicting the test results for linear regression

```python
y_pred=LogisticRegression.predict(x_test)
```

```python
result=pd.DataFrame(np.c_[y_test,y_pred],columns=["Original","Predicted"])
result
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.metrics import accuracy_score,classification_report
cs=classification_report(y_test,y_pred)
print(cs)
```
