Clusters Topic we form the grouping or clusters

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("J:\Mall_Customers.csv")
df.head()
```

```python
plt.scatter(df["Spending Score (1-100)" ],df["Annual Income (k$)"])
plt.xlabel("Spending Score (1-100)")
plt.ylabel("Annual Income (k$)")
plt.title("Mall Customers")
```

```python
relevant_cols=["Age","Annual Income (k$)",
               "Spending Score (1-100)"]
customer_df=df[relevant_cols]
```

Standard Scalar

```python
from sklearn.preprocessing  import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(customer_df)
```

```python
scaled_data=scaler.transform(customer_df)
```

```python
scaled_data
```

To Form The clusters

```python
def find_best_clusters(df,maximum_K):
    clusters_centers=[]
    k_values=[]  # First k values are stored in this
    for k in range(2,maximum_K): # IN this we create two clusters for any data
        kmeans_model=KMeans(n_clusters=k)
        kmeans_model.fit(df)

        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)
    return clusters_centers,k_values
```

```python
clusters_centers,k_values=find_best_clusters(scaled_data,12)
```

```python
clusters_centers
```

```python
k_values
```

```python
def generate_elbow_plot(clusters_centers,k_values):

    figure=plt.subplots(figsize=(12,6))
    plt.plot(k_values,clusters_centers,'o-',color='blue')
    plt.xlabel("Number of clusters (k)")
    plt.ylabel("Clusters inertia")
    plt.title("Elbow Plot of KMeans ")
    plt.show()
```

```python
generate_elbow_plot(clusters_centers,k_values)
```
