```python
import pandas as pd,numpy as np
df= pd.DataFrame({'age':[20, 20, 21, 21,22,22],
                  'skill rating':[3, 4, 2, 3,3,5],
                  'Salary_type':[1,1,0,0,0,1]})
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
df.age=sc.fit_transform(np.array(df.age).reshape(-1, 1))
```

```python
X = df.iloc[:,:-1]

y = df.Salary_type

y
```

```python
data = np.array(X)

labels = np.array(y)
data.shape[1]
data
```

```python
np.random.seed(42) # seed for reliable values
n_classes = len(np.unique(labels)) # number of classes

weights = np.random.rand(2, 2+1) # weights initial values

epochs = 50

for epoch in range(epochs):

 correct_classifications = 0

 for row in range(data.shape[0]):

  data_with_bias = np.insert(data[row], 0, 1)

  z = np.dot(weights,data_with_bias)

  pred_class = np.argmax(z)

  error = labels[row] - pred_class

  if error == 0:

   correct_classifications += 1

  else:

   weights[pred_class] -= (0.01*data_with_bias)

   weights[labels[row]] += (0.01*data_with_bias)

 acc = correct_classifications / data.shape[0]

 print(f'Epoch {epoch+1}: Accuracy = {acc}')
```

```python
new_value = [1,0, 5]

z = np.dot(weights,new_value)

output = int(np.argmax(z))

output
```
