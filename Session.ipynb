```python
import numpy as np
import pandas as pd
from keras import Sequential
from tensorflow.keras.layers import Convolution2D,MaxPooling2D  #max pooling layer it is a 2d
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data = ImageDataGenerator(1./255,zoom_range=0.2) 
```

```python
data1 = data.flow_from_directory(r'C:\Users\HP\Downloads\archive (3)\train',target_size=(64,64))
```

Architecture


```python
model=Sequential()
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3)))#64 rows,64 columns and 3 color channels
model.add(MaxPooling2D(2,2))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(Convolution2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(66,activation='leaky_relu'))
model.add(Dense(33,activation='tanh'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```
