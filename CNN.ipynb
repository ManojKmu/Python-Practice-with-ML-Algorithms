First Neural Network with keras  (Deep Learning) we have tensorflow packages

```python
from numpy import loadtxt
from tensorflow import keras

from tensorflow.keras.models import Sequential  # Builded Architecure is stored in the Sequential class
from tensorflow.keras.layers import Dense  # Using dense we create the hidden layer
```

```python
# load the dataset
dataset = loadtxt('J:\pima-indians-diabetes.csv', delimiter=',') # loadtxt is used to sepereate the data with comma
```

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8] # It is used to drop a particular column using slicing
y = dataset[:,8]
```

```python
# define the keras model (This is the actual deep learning architecture)
model = Sequential()
model.add(Dense(8, input_dim=8, activation='relu')) #input layer (Based on the dataset we decide the input_dimension)
model.add(Dense(69, activation='relu')) #hidden layer we take the neuron value based on accuracy 
model.add(Dense(1, activation='sigmoid')) #output layer   we use sigmoid for binary classification to get a single output 0 or 1
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=30, batch_size=5)  # epochs is used to define the no.of forward and backward propagation running in background 
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

# Now let us add additional hidden layers into our network and observe

```python
#Adding new layers
model_new = Sequential()
model_new.add(Dense(8, input_dim=8, activation='relu')) #input layer
model_new.add(Dense(64, activation='relu')) # 1st hidden layer with 64 neurons
model_new.add(Dense(50, activation='softmax')) # 2nd hidden layer with 32 neurons
model_new.add(Dense(70, activation='linear')) # 3rd hidden layer with 16 neurons
model_new.add(Dense(1, activation='sigmoid')) #output layer
```

```python
# compile the keras model
model_new.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])#Identify the loss function we use 'binary_crossentropy' for binary classification(sigmoid activation function)
```

```python
# fit the modified keras model on the dataset
model_new.fit(X, y, epochs=20, batch_size=5)
```

```python
# evaluate the keras model
_, accuracy = model_new.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
